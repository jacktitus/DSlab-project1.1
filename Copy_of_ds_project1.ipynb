# 3. Find missing values and impute appropriately(mean/median/mode)

```python
missing_values = df.isnull().sum()
print(missing_values)
```

# 4. Compute summary statistics

```python
summary_stats = df.describe()
print(summary_stats)
```

```python
for col in df.select_dtypes(include=['float', 'int']).columns:
    df[col] = df[col].fillna(df[col].mean())
for col in df.select_dtypes(include='object').columns:
  df[col] = df[col].fillna(df[col].mode())

```

# 5. Sort dataset by one column


```python
sort_by_price = df.sort_values(by = 'price', ascending= False )
print(sort_by_price)
```

# 6. filter records with conditions

```python
filtered = df[(df['bedrooms'] >= 3) & (df['price'] < 500000)]
filtered
```


# 7. Derived column

#  9. Outlier detection (Z-score)

```python
from scipy.stats import zscore
outliers = (abs(zscore(df['price'])) > 3)
df_no_outliers = df.loc[~outliers]
```

```python
import matplotlib.pyplot as plt
plt.boxplot(df['price'])
plt.title('Price Before Outlier Treatment')
plt.show()
```

```python
plt.boxplot(df_no_outliers['price'])
plt.title('Price After Outlier Treatment')
plt.show()
```
